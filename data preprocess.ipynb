```python
from math import sin
from math import cos
from math import atan2
from math import sqrt

def get_distance(lat1,lon1,lat2,lon2):
    R = 6370
    dlon = lon2 - lon1
    dlat = lat2- lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

df['distance_ori']   = df.apply(lambda x: get_distance(x.origin_lat, x.origin_lon,x.event_lat, x.event_lng) ,axis=1)
df['distance_des']   = df.apply(lambda x: get_distance(x.des_lat, x.des_lon,x.event_lat, x.event_lng) ,axis=1)
```

```python
from datetime import datetime
from dateutil import tz
```

```python
from_zone = tz.gettz('UTC')

def get_local_tz(timezone,utc):
    to_zone = tz.gettz(timezone)
    utc = utc.replace(tzinfo=from_zone)
    return utc.astimezone(to_zone)
```

```python
def get_format_deliver_time(x):
    try:
        return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    except:
        return datetime.strptime(x[:10], '%Y-%m-%d')

df['format_delivery_date_time_utc']=df['delivery_date_time_utc'].apply(get_format_deliver_time)
```

```python
df['format_delivery_date_time_local']   = df.apply(lambda x: get_local_tz(x.event_time_zone, x.format_delivery_date_time_utc) ,axis=1)
```

```python
def get_timestamp(x):
    
    t=x.split('.')[0]
    #try:
    return datetime.strptime(t, '%Y-%m-%d %H:%M:%S')
    #except:
      #  return datetime(2022, 12, 31, 0, 0)
    
df['format_event_date_time_local']=df['event_date_time_local'].apply(get_timestamp)
```

```python
df['format_event_date_time_local']=df['format_event_date_time_local'].apply(lambda x: x.replace(tzinfo=None))
```

```python
df['delivery_period'] =  df['format_delivery_date_time_local'].apply(lambda x:x.day)-df['format_event_date_time_local'].apply(lambda x:x.day)
```

```python
import time
import datetime
import math
import numpy as np


df['pracel_month']=df['format_event_date_time_local'].apply(lambda x:x.month)

df["month_norm"] = 2 * math.pi * df['pracel_month'] / df['pracel_month'].max()

df["cos_month"] = np.cos(df['month_norm'])
df["sin_month"] = np.sin(df['month_norm'])
```

```python
df['pracel_week']=df['format_event_date_time_local'].apply((lambda x:x.weekday()))

df["week_norm"] = 2 * math.pi * df['pracel_week'] / df['pracel_week'].max()

df["cos_week"] = np.cos(df['week_norm'])
df["sin_week"] = np.sin(df['week_norm'])
```

```python
# Create a function that calculates the exponential decay
from datetime import datetime, timedelta
def decay_factor(decayInDays, offset,days_decay):

   # factor = np.exp(-decayInDays / days_decay)
    if decayInDays <=(30+offset):
        factor=1
    elif decayInDays < (45+offset):
        factor =0.8
    elif decayInDays < (60+offset):
        factor = 0.6
    elif decayInDays <(90+offset):
        factor =0.5
    elif decayInDays < (180+offset):
        factor = 0.4
    elif decayInDays < (400+offset):
        factor =0.3
    else:
        factor=0.15
        
    return factor
```

```python
from sklearn.metrics import mean_pinball_loss

#Set the minimum error arbitrarily large
min = 99999999999999999999999 
count = 0 #Used for keeping track of the iteration number
#How many runs to perform using randomly selected hyperparameters
iterations = 20
alphaVal=0.8
for i in range(iterations):
    print('iteration number', count)
    count += 1 #increment count
    try:
        
        d_train = lgbm.Dataset(X_train, label=Y_train,weight=decay_rate_weight) #Load in data
        params = {} #initialize parameters
        params['learning_rate'] = np.random.uniform(0, 1)
        params['boosting_type'] = np.random.choice(['gbdt', 'dart', 'goss'])
        params['objective'] = 'quantile'
        params['alpha'] = alphaVal
        params['metric'] = 'quantile'
        params['sub_feature'] = np.random.uniform(0, 1)
        params['num_leaves'] = np.random.randint(20, 300)
        params['min_data'] = np.random.randint(10, 100)
        params['max_depth'] = np.random.randint(5, 200)
        params['feature_pre_filter'] = 'false'
        iterations = np.random.randint(10, 500)
        print(params, iterations)
#Train using selected parameters
        clf = lgbm.train(params, d_train, iterations)
        y_pred=clf.predict(X_test) #Create predictions on test set
        mae=mean_absolute_error(y_pred,Y_test)
        print('MAE:', mae)
        loss = mean_pinball_loss(Y_test, y_pred,alpha= alphaVal)
        print('Pinball loss:', loss)
        if mae < min:
            min = mae
            pp = params 
    except Exception as e: #in case something goes wrong
        print('failed with')
        print(e)
print("*" * 50)
print('Minimum is: ', min)
print('Used params', pp)
```

```python
# prediction
#Y_pred = clf.predict(X_test)
def printMetrics(y_pred,Y_test):
    # Error Metrics for Final Model with sample of (1,000,000) of dataset
    # Calculate MAPE
    y_test, y_pred = np.array(Y_test), np.array(y_pred)

    MAPE = 0
    for i in range(y_test.size):
        MAPE = MAPE + ((abs(y_test[i] - y_pred[i]))/ (abs(y_test[i])))
    MAPE = (MAPE / (abs(y_test.size)))*100
    print('MAPE = : ',MAPE) 


    # Calculate MAE
    MAE = 0
    for i in range(y_test.size):
        MAE = MAE + (abs(y_test[i] - y_pred[i]))
    MAE = (MAE / (abs(y_test.size)))
    print('MAE = ', MAE)


    # Calculate MSE
    MSE=0
    for i in range(y_test.size):
        MSE = MSE + ((abs(y_test[i] - y_pred[i]))**2)
    MSE = (MSE / (abs(y_test.size)))

    # Calculate RMSE
    RMSE = np.sqrt(MSE)
    print('RMSE = ',RMSE)

    # Calculate MEDIAN_AE
    from sklearn.metrics import median_absolute_error
    MEDIAN_AE = median_absolute_error(y_test, y_pred)
    print('MEDIAN_AE = ',MEDIAN_AE)


    # Calculate R2_Score
    from sklearn.metrics import r2_score
    R2_Score = r2_score(y_test, y_pred)
    print('R2_Score = ',R2_Score)
    

print("**************************************")
printMetrics(y_pred,Y_test)
```
